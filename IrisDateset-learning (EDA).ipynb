```python
#learning of Iris Data Set Classification

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
iris=pd.read_csv("datasets/iris.csv")
```

```python
iris["species"].value_counts()
```

```python
iris.plot(kind="scatter",x="sepal_length",y="sepal_width")
```

```python
#analysing on the basis of sepal length
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="species",height=4)\
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend();
```

```python
#analysing on the basis of petal length
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="species",height=4)\
.map(plt.scatter,"petal_length","petal_width")\
.add_legend();
```

```python
#pair plot
sns.set_style("whitegrid")
sns.pairplot(iris,hue="species",height=3)
plt.show()
```

```python
#univariate analysis

#distribution on Petal Length
sns.FacetGrid(iris,hue="species",height=5)\
.map(sns.distplot,"petal_length")\
.add_legend()

```

```python
#distribution on Petal width
sns.FacetGrid(iris,hue="species",height=5)\
.map(sns.distplot,"petal_width")\
.add_legend()
```

```python
#distribution on sepal Length
sns.FacetGrid(iris,hue="species",height=5)\
.map(sns.distplot,"sepal_length")\
.add_legend()
```

```python
#distribution on sepal width
sns.FacetGrid(iris,hue="species",height=5)\
.map(sns.distplot,"sepal_width")\
.add_legend()
```

```python
np.median(iris[iris["species"]=="setosa"]["petal_length"])
```

```python
#box plot- to understand percentiles at 25, 50 and 75 percentile
sns.boxplot(x="species",y="petal_length",data=iris)
```

```python
#violin plot- combines both world of histogram and pdf

sns.violinplot(x="species",y="petal_length",data=iris,size=8)
```

```python
#2D density plot using joint plot
sns.jointplot(x="petal_length",y="petal_width",data=iris[iris["species"]=="setosa"],kind="kde")
```
