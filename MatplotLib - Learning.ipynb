```python
#matplot lib to generate scatter graph
import matplotlib.pyplot as pyplot
#data

riding=([17, 18, 21, 22, 19, 21, 25, 22, 25, 24],[3, 6, 3.5, 4, 5, 6.3, 4.5, 5, 4.5, 4])
swimming = ((17, 18, 20, 19, 22, 21, 23, 19, 21, 24),(8, 9, 7, 10, 7.5, 9, 8, 7, 8.5, 9)) 
sailing = ((31, 28, 29, 36, 27, 32, 34, 35, 33, 39),(4, 6.3, 6, 3, 5, 7.5, 2, 5, 7, 4))


pyplot.scatter(x=riding[0],y=riding[1],c='red', marker='o',label='riding')
pyplot.scatter(x=swimming[0], y=swimming[1], c='green', marker='^', label='swimming') 
pyplot.scatter(x=sailing[0], y=sailing[1], c='blue', marker='*', label='sailing')



pyplot.xlabel('Age')
pyplot.ylabel('Hours')

pyplot.title('Activities Scatter Graph')
pyplot.legend()
pyplot.show()
```

```python
#generating trend line with Scatter Plot
import numpy as np
import matplotlib.pyplot as pyplot
x=(5, 5.5, 6, 6.5, 7, 8, 9, 10)
y=(120, 115, 100, 112, 80, 85, 69, 65)

pyplot.scatter(x,y,c='red')

#pyplot.show()

#generate the trend line
z=np.polyfit(x,y,1)
p=np.poly1d(z)

pyplot.plot(x,p(x),'r')

pyplot.show()
```

```python
#generating pie charts using matplotlib
#A Pie Chart is a type of graph in which a circle is divided into sectors (or wedges) that each represent a proportion 
#of the whole. A wedge of the circle represents a category’s contribution to the overall total. As such the graph 
#resembles a pie that has been cut into different sized slices.

import matplotlib.pyplot as pyplot
labels=('Python','Java','Scala','C#')
sizes=[45,30,15,10]

pyplot.pie(sizes,labels=labels,autopct='%1.f%%',counterclock=False,startangle=90,explode=(.1,0,0,0),shadow=True)
pyplot.show()

```

```python
#Bar Chart- Generating Bar Charts using matplotlib

#A Bar Chart is a type of chart or graph that is used to present different discrete categories of data. 
#The data is usually presented vertically although in some cases horizontal bar charts may be used. 
#Each category is represented by a bar whose height (or length) represents the data for that category
import matplotlib.pyplot as pyplot
labels=('Python','Scala','C#','Java','PHP')
index=(1,2,3,4,5)
values=[45,10,15,30,22]
colors=['red','blue','green','black','cyan']

pyplot.bar(index,values,tick_label=labels,color=colors)
pyplot.show()
pyplot.barh(index,values,tick_label=labels,color=colors)
pyplot.show()


```

```python
#Stacked Bar Chart

import matplotlib.pyplot as pyplot
labels=('Python', 'Scala', 'C#', 'Java', 'PHP') 
index=(1,2,3,4,5)
web_usage=[20, 2, 5, 10, 14]
data_science_usage=[15, 8, 5, 15, 2] 
games_usage=[10, 1, 5, 5, 4]

pyplot.bar(index,web_usage,tick_label=labels,label='web')
pyplot.bar(index,data_science_usage,tick_label=labels,label='data science',bottom=web_usage)

web_and_games_usage = [web_usage[i] + data_science_usage[i] for i in range(0, len(web_usage))] 

pyplot.bar(index, games_usage, tick_label=labels, label='games', bottom=web_and_games_usage)

pyplot.ylabel('Usage') 
pyplot.xlabel('Programming Languages') 
pyplot.legend()


pyplot.show()
```

```python
#3D Graphs
#A three dimensional graph is used to plot the relationships between three sets of values 
#(instead of the two used in the examples presented so far in this chapter). 
#In a three dimensional graph as well as the x and y axis there is also a z axis. 
import matplotlib.pyplot as pyplot
from matplotlib import cm as colourmap
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x_values = np.arange(-6, 6, 0.3) 
y_values = np.arange(-6, 6, 0.3)

x_values, y_values = np.meshgrid(x_values, y_values)
z_values = np.sin(x_values + y_values)
figure = pyplot.figure()
axes = figure.gca(projection='3d')

surf = axes.plot_surface(x_values,y_values,z_values,cmap=colourmap.coolwarm)
figure.colorbar(surf)
pyplot.title("3D Graph") 
axes.set_ylabel('y values', fontsize=8) 
axes.set_xlabel('x values', fontsize=8) 
axes.set_zlabel('z values', fontsize=8)

pyplot.show()
```
